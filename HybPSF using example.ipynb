```python
from HybPSF import _Image2PSFmodel,_get_webPSF
import numpy as np
import matplotlib.pyplot as plt
#full path of fitsfile
fname="/Users/linn/Documents/soft/HybPSF/tests/jw02736001001_02105_00009_nrcalong_cal.fits"
#path to write the calculated files
wdirs="/Users/linn/Documents/soft/HybPSF/tests/"
#fitting star images for normal PSF
PCname=_Image2PSFmodel(fname,wdirs,NPCs=5,SNR_thresh=10.,osam=4)
print((PCname)) #PSF model path
```

```python
#given the target PSF position
pos=np.array([[0,0]]) #interpolated PSF position (pixel)
#get the intepolated PSF model using _get_webPSF
rPSFs=_get_webPSF(PCname,pos,25) #rPSFs[id][x-direction][y-direction]
plt.imshow(rPSFs[0]**0.1)
plt.show()
```
